```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("CooperUnion/anime-recommendations-database")

print("Path to dataset files:", path)
```

```python
import numpy as np
import pandas as pd
```

```python
animes = pd.read_csv(f'{path}/anime.csv')
```

```python
animes.head()
```

```python
animes = animes[["anime_id", "name", "genre", "type", "episodes"]]
```

```python
animes.head()
```

```python
animes.shape
```

```python
animes.info()
```

```python
animes.dropna(inplace=True)
```

```python
animes.shape
```

```python
animes.duplicated().sum()
```

```python
animes['genre'].dtype
```

```python
type(animes['genre'].iloc[0])
```

```python
animes['genre'] = animes['genre'].str.split(", ").apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
animes['type'] = animes['type'].str.split(" ")
animes['episodes'] = animes['episodes'].str.split(" ")
```

```python
animes['type']
```

```python
animes['title'] = animes['name']
```

```python
animes['tag'] = animes['genre'] + animes['type'] + animes['episodes']
```

```python
df = animes[['anime_id', 'title' , 'tag']]
```

```python
df.head()
```

```python
df['tag'] = df['tag'].str.join(" ").str.lower()
df['tag']
```

```python
df['tag'][0]
```

```python
# covert each word to vectors
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000)
```

```python
vectors = cv.fit_transform(df['tag']).toarray() # converting scipy matrix to array
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# 0 --> not similar
# 1 --> similar
simil = cosine_similarity(vectors)
```

```python
simil
```

```python
simil.shape
```

```python
sorted(list(enumerate(simil[0])), key = lambda x: x[1], reverse=True)[1:6]
```

```python
def recommend(anime):
  anime_idx = df[df['title'] == anime].index[0]
  dist = simil[anime_idx]
  anime_list = sorted(list(enumerate(dist)), key = lambda x: x[1], reverse=True)[1:6]
  for i in anime_list:
    print(df.iloc[i[0]].title)
```

df[df['name'] == 'Steins;Gate']

```python
recommend('Steins;Gate')
```

```python
recommend('Gintama°')
```
